```python
import numpy as np
import pandas as pd
import seaborn as srn
from matplotlib import pyplot as plt
```

```python
table = pd.read_csv('metrics.csv')
table
```

```python
bfs = table.loc[table['algorithm'] == 'bfs']
bfsF = bfs.loc[table['difficulty'] == 'F']
bfsF10 = bfsF.loc[table['dim'] == 10]
bfsF20 = bfsF.loc[table['dim'] == 20]
bfsF50 = bfsF.loc[table['dim'] == 50]

bfs = table.loc[table['algorithm'] == 'bfs']
bfsM = bfs.loc[table['difficulty'] == 'M']
bfsM10 = bfsM.loc[table['dim'] == 10]
bfsM20 = bfsM.loc[table['dim'] == 20]
bfsM50 = bfsM.loc[table['dim'] == 50]

bfs = table.loc[table['algorithm'] == 'bfs']
bfsD = bfs.loc[table['difficulty'] == 'D']
bfsD10 = bfsD.loc[table['dim'] == 10]
bfsD20 = bfsD.loc[table['dim'] == 20]
bfsD50 = bfsD.loc[table['dim'] == 50]

kohonen = table.loc[table['algorithm'] == 'kohonen']
kohonenF = kohonen.loc[table['difficulty'] == 'F']
kohonenF10 = kohonenF.loc[table['dim'] == 10]
kohonenF20 = kohonenF.loc[table['dim'] == 20]
kohonenF50 = kohonenF.loc[table['dim'] == 50]

kohonen = table.loc[table['algorithm'] == 'kohonen']
kohonenM = kohonen.loc[table['difficulty'] == 'M']
kohonenM10 = kohonenM.loc[table['dim'] == 10]
kohonenM20 = kohonenM.loc[table['dim'] == 20]
kohonenM50 = kohonenM.loc[table['dim'] == 50]

kohonen = table.loc[table['algorithm'] == 'kohonen']
kohonenD = kohonen.loc[table['difficulty'] == 'D']
kohonenD10 = kohonenD.loc[table['dim'] == 10]
kohonenD20 = kohonenD.loc[table['dim'] == 20]
kohonenD50 = kohonenD.loc[table['dim'] == 50]

prog_din = table.loc[table['algorithm'] == 'prog_din']
prog_dinF = prog_din.loc[table['difficulty'] == 'F']
prog_dinF10 = prog_dinF.loc[table['dim'] == 10]
prog_dinF20 = prog_dinF.loc[table['dim'] == 20]
prog_dinF50 = prog_dinF.loc[table['dim'] == 50]

prog_din = table.loc[table['algorithm'] == 'prog_din']
prog_dinM = prog_din.loc[table['difficulty'] == 'M']
prog_dinM10 = prog_dinM.loc[table['dim'] == 10]
prog_dinM20 = prog_dinM.loc[table['dim'] == 20]
prog_dinM50 = prog_dinM.loc[table['dim'] == 50]

prog_din = table.loc[table['algorithm'] == 'prog_din']
prog_dinD = prog_din.loc[table['difficulty'] == 'D']
prog_dinD10 = prog_dinD.loc[table['dim'] == 10]
prog_dinD20 = prog_dinD.loc[table['dim'] == 20]
prog_dinD50 = prog_dinD.loc[table['dim'] == 50]
```
